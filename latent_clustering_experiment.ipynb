```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.mixture import GaussianMixture
```

```python
normalized_history = pd.read_csv('normalized_history_data.csv')
normalized_history
```

```python
cov_types = ['full', 'tied', 'diag', 'spherical']
latent_estimators = [GaussianMixture(n_components=15, covariance_type=cov_type).fit(normalized_history) for cov_type in cov_types]

n_cov_types = np.arange(len(cov_types))

plt.figure()
plt.title('Latent Estimator AIC Value')
plt.bar(n_cov_types, [latent_estimator.aic(normalized_history) for latent_estimator in latent_estimators])
plt.xlabel('Covariance Type')
plt.ylabel('AIC')
plt.xticks(n_cov_types, cov_types, rotation=30)
plt.show()

plt.figure()
plt.title('Latent Estimator BIC Value')
plt.bar(n_cov_types, [latent_estimator.bic(normalized_history) for latent_estimator in latent_estimators])
plt.xlabel('Covariance Type')
plt.ylabel('BIC')
plt.xticks(n_cov_types, cov_types, rotation=30)
plt.show()
```

Nilai AIC dan BIC lebih baik apabila nilainya lebih kecil.

Berdasar grafik di atas, didapat bahwa nilai AIC dan BIC paling kecil ketika covariance type yang digunakan adalah `tied`. Oleh karena itu, clustering variabel laten akan menggunakan covariance type `tied`.
